## Dataset Description: 

The Iris dataset is a well-known dataset in the field of machine learning and statistics. It contains 150 samples of iris flowers from three species: Iris setosa, Iris versicolor, and Iris virginica.

   
## Features: Four numerical features representing physical measurements of the flowers

Sepal Length (in cm)

Sepal Width (in cm)

Petal Length (in cm)

Petal Width (in cm)

## Classes: The species of the iris flower (categorical variable).

Setosa: 50 instances

Versicolor: 50 instances

Virginica: 50 instances

Balanced: Each class has an equal number of instances (50).

Simple and Small: Suitable for beginners in machine learning.

Well-Studied: Commonly used for benchmarking algorithms.

```python
# Handling Outliers

sns.boxplot(x='sepalwidth', data=df)
```

```python
# Function to remove outliers based on IQR
def remove_outliers(df, column):
    Q1 = np.percentile(df[column], 25, interpolation='midpoint')
    Q3 = np.percentile(df[column], 75, interpolation='midpoint')
    IQR = Q3 - Q1

    # Upper and lower bounds
    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Filtering outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

print("Old Shape:", df.shape)

# Removing outliers from 'sepalwidth' column
df = remove_outliers(df, 'sepalwidth')

print("New Shape:", df.shape)

# Plotting boxplot after removing outliers
sns.boxplot(x='sepalwidth', data=df)
```
